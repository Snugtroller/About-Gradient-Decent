```python
import pandas as pd
import numpy as np
data=pd.read_csv("/content/Salary_Data.csv")
```

```python
X=data.iloc[:,:-1].values
y=data.iloc[:,1].values
print(X)
print(y)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X,y,color="blue")
plt.xlabel("YearsExperience")
plt.ylabel("Salary")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(X,y)
print(lr.coef_)
print(lr.intercept_)
```

```python
plt.scatter(X,y,color="blue")
plt.plot(X,lr.predict(X),color="red")
plt.xlabel("YearsExperience")
plt.ylabel("Salary")
plt.show()
```

This is the line when our intercept is 0


```python
y_pred=9449.96*X+0
plt.scatter(X,y)
plt.plot(X,lr.predict(X),color="red")
plt.plot(X,y_pred,color="blue")
plt.show()
```

```python
c=0
m=9449.96
learning_rate=0.01
lost_slope=-2*np.sum(y-m*X.ravel()-c)
step=lost_slope*learning_rate
c=c-step
y_pred1=9449.96*X+c
```

Now the line will tend to move upwards as mentioned and move to the given intercept and if we keep iterating it will go on moving up.


```python
plt.scatter(X,y)
plt.plot(X,lr.predict(X),color="red")
plt.plot(X,y_pred,color="blue")
plt.plot(X,y_pred1,color="green")
```

*Now we take the coeffecient that is the slope which will remain the constant throughout and keep moving the intercept till we get the required intercept that is 25792.200*

```python
c=0
m=9449.96
iteration=100
learning_rate=0.01
for i in range(iteration):
  lost_slope=-2*np.sum(y-m*X.ravel()-c)
  c=c-(learning_rate*lost_slope)
  y_pred=m*X+c
  plt.plot(X,y_pred)
plt.scatter(X,y)
```

For this model we are getting an accuracy of 95%


```python
from sklearn.metrics import r2_score
r2_score(y,y_pred)
```
